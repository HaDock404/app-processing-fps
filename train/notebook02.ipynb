```python
import cv2
import os

# === Paramètres ===
input_folder = "./video_processed_1"
output_folder = "images_redim_1"
new_width, new_height = 2048, 1024

os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith((".jpg", ".png")):
        image_path = os.path.join(input_folder, filename)
        image = cv2.imread(image_path)

        if image is None:
            print(f"❌ Erreur lecture image : {filename}")
            continue

        resized_image = cv2.resize(image, (new_width, new_height), interpolation=cv2.INTER_AREA)

        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, resized_image)
        print(f"✅ Image redimensionnée : {filename}")

print(f"\n📁 Toutes les images ont été redimensionnées et enregistrées dans : {output_folder}")
```
